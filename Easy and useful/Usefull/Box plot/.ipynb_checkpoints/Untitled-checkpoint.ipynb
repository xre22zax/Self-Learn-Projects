```python
import pandas as pd
from matplotlib import pyplot as plt

healthcare = pd.read_csv("healthcare.csv")
print(healthcare.head())
```

We’re going to focus on only the ways in which hospitals charge patients for chest pain. 

```python
DRG = healthcare["DRG Definition"].unique()
```

```python
chest_pain = healthcare[healthcare['DRG Definition'] == '313 - CHEST PAIN']
```

Let’s make a boxplot for every state!
To begin, we first need to create a list of all the states in our dataset. We can once again use the unique() function.

```python
states = chest_pain['Provider State'].unique()
```

```python
datasets = []
for state in states:
  datasets.append(chest_pain[chest_pain['Provider State'] == state][' Average Covered Charges '].values)
```

We’re about to draw 50 boxplots.

```python
plt.figure(figsize=(20,6))
plt.boxplot(datasets, labels = states )
plt.show()
```
